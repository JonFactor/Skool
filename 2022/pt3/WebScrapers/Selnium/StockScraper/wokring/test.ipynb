```python
# get data
data = []
for file in os.listdir('D:\CodeShit\\try'):
    with open(f'D:\CodeShit\\try\\{file}', 'r') as read:
        data.append(json.load(read))
```

```python
# find ids and place id
x2 = 0
for DICT in data:
    info = DICT['values']
    id = DICT['cik'].replace('[', '').replace('].json', '')
    # get lowest
    Dates = []
    for bundle in info:
        for elemen in bundle: 
            try: 
                if '.' not in str(elemen[0][1]):
                    Dates.append(str(elemen[0][1].replace('-', '')))
            except Exception: pass
    lowestDate = min(Dates)
    tick = yf.Ticker(id)
    years = 2023 - int(lowestDate[:4]) + 1
    history = tick.history(f'{years}y', '1d')['Open']
    x = 0
    for bundles in info:
        for elements in bundles:
            if not len(elements) < 1:
                date = str(elements[0][1].replace('-', '/'))
                if not len(str(date)) < 9 and '.' not in str(date):
                    dateformated = dt.datetime.strptime(date, '%Y/%m/%d')
                    end = dateformated + timedelta(days=7)
                    end = history.iloc[np.argmin(np.abs(history.index.to_pydatetime() - end.replace(tzinfo=dt.timezone.utc)))]
                    start = history.iloc[np.argmin(np.abs(history.index.to_pydatetime() - dateformated.replace(tzinfo=dt.timezone.utc)))]
                    if start - end < 0 :
                        y = True
                    else: y = False
                    elements.append(y)

                else:
                    info[x].remove(elements)
            else:
                info[x].remove(elements)
        x += 1
            
   #     except Exception: 
  #          data[x2]['values'][x].append('NOPE')
 #           x += 1
    x2 += 1
```
